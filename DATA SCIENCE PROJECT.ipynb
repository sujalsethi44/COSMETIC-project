```python
import numpy as np
import matplotlib.pyplot as  plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv('cosmetics.csv')
df["Price"].fillna(144, inplace=True)                 #empty value fill up
df["Rank"].fillna(5, inplace=True)
```

```python
display(df.sample(10))           #different type sample
```

```python
df.plot()
plt.show()
```

```python
df.plot(kind='scatter', x= 'Brand', y='Price')
plt.show()
```

```python
ax = df.plot(x="Brand", y="Oily", kind="bar")
df.plot(x="Brand", y="Dry", kind="bar", ax=ax, color="C2")
df.plot(x="Brand", y="Sensitive", kind="bar", ax=ax, color="C3")
ax.legend(fontsize = 15)  #oily, sensitive font size mention
plt.show()
```
